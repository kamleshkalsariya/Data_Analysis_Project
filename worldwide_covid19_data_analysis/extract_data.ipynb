####  Covid-19 Data Source: https://covid19.who.int/info/

```python
import pandas as pd
```

```python
#read csv file from url 
url1 = 'https://covid19.who.int/WHO-COVID-19-global-data.csv'
df = pd.read_csv(url1)
df.head()
```

```python
#To store csv file into local machine
df.to_csv("WHO_COVID-19_global_regular_data.csv")
```

```python
df1 = pd.read_csv("https://covid19.who.int/WHO-COVID-19-global-table-data.csv",index_col=False)
df1.head()
```

```python
#extract total covid-19 global data
df2 = df1[df1['WHO Region'].isna()]
df2.head()
```

```python
#To store file
df2.to_csv("WHO_COVID_19_global_total_data.csv")
```

```python
#extract countrywise total covid-19  data
df3 = df1[~df1['WHO Region'].isna()]
df3.head()
```

```python
#To store file
df3.to_csv("WHO_COVID_19_Country_total_data.csv")
```

```python
#read vaccination global data
df4 = pd.read_csv('https://covid19.who.int/who-data/vaccination-data.csv')
df4.head()
```

```python
#To store vaccination file
df4.to_csv("WHO_COVID_19_Country_vaccination_data.csv")
```
